```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
%matplotlib inline
import sklearn
import scipy
import seaborn
```

Clustering Dataset used is Blobs 

```python
from sklearn.datasets import make_blobs
```

```python
X, y = make_blobs(n_samples=10, centers=3, n_features=2, cluster_std=0.40, random_state=42)
print(X.shape)
print(y.shape)
```

```python
y
```

First 5 values in X

```python
X[:5]
```

```python
plt.scatter(X[:,0], X[:,1], s=50)
```

K Means Clustering

```python
from sklearn.cluster import KMeans 

kmeans = KMeans(n_clusters=3, verbose=1) 
kmeans.fit(X)
y_kmeans = kmeans.predict(X)
```

```python
centers = kmeans.cluster_centers_
print(centers)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')
plt.scatter(centers[:, 0], centers[:, 1], c='red' , s=200, alpha=0.5)
```

B] Hierarchical Clustering


```python
from scipy.cluster.hierarchy import linkage, dendrogram
Z1 = linkage(X, method='single', metric='euclidean') #single
```

```python
ticks = ['el_{}_group={}'.format(i, el) for i, el in enumerate(y)]
```

```python
plt.figure(figsize=(25, 10))
plt.title('Hierarchical Clustering')
plt.xlabel('samples')
plt.ylabel('distance')
dendrogram(Z1)

plt.show()
```

```python
Z2 = linkage(X, method='complete', metric='euclidean') #complete
```

```python
plt.figure(figsize=(25, 10))
plt.title('Hierarchical Clustering')
plt.xlabel('samples')
plt.ylabel('distance')
dendrogram(Z2)

plt.show()
```

```python
Z3 = linkage(X, method='average', metric='euclidean') #average
```

```python
plt.figure(figsize=(25, 10))
plt.title('Hierarchical Clustering')
plt.xlabel('samples')
plt.ylabel('distance')
dendrogram(Z3)

plt.show()
```
