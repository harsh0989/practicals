```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('weight-height.csv')
df.head()
```

```python
df.info()
```

```python
df = df.drop('Gender', axis = 1)
df.info()
```

```python
X = df.values[:,0] 
y = df.values[:,1]
m = len(X) 
X = X.reshape(m, 1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=0)
print(X_train)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, y_train)
print(f'Training accuracy: {lr.score(X_train, y_train)}')
print(f'Testing accuracy: {lr.score(X_test, y_test)}')
```

```python
print(lr.coef_)
print(lr.intercept_)
```

```python
plt.plot(X, y, '.')
X_line = np.linspace(50, 80, num = 100)
y_line = lr.coef_*X_line + lr.intercept_
plt.plot(X_line, y_line)
```

```python
from sklearn.metrics import mean_squared_error

predictions = lr.predict(X_train)
print(f"Mean squared error: {mean_squared_error(y_train, predictions)}")
```
